Now we want to store all the edges of the graph and we define a helper function for the same.

```python
def gen_graph(graph):
    # define the list to store all the edges in the graph
    edges = []
    for node in graph:
        for neighbour in graph[node]:
            # appending the list with tuples of edges connected...
            edges.append((node, neighbour))
    return edges

def gen_iso_graph(graph):
    # define the list to store all the edges in the graph
    edges_iso = []
    for node in graph:
        if not graph[node]:
            # appending the list of isolated edges...
            edges_iso.append(node)
    return edges_iso
```

```python
print(gen_graph(graph))
```

```python
print(gen_iso_graph(graph)) # obtains the isolated node from the graph
```
